## Create simple model

```python
model = BasicMLModel(
    num_humans=50,
    num_ml=15,
    belief_dims=30,
    p_1=0.1,
    p_2=0.9,
    p_3=0.9,
    p_hp=0.1,
    p_hm=0.1,
    p_ml=0.5,
)
```

```python
def print_model(model):
    print("MODEL CONFIGURATION")
    print(model.conf)
    print("MODEL AGENTS")
    for a in model.schedule.agents:
        print(f"{a.unique_id}, kl={a.kl}, state={a.state}")
```

```python
print_model(model)
```

```python
for _ in range(100):
    model.step()
print_model(model)
```
